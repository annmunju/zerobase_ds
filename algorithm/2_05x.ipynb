```python
mykor=int(input("국어 점수: "))
myeng=int(input("영어 점수: "))
mymath=int(input("수학 점수: "))
mysce=int(input("과학 점수: "))
myhis=int(input("역사 점수: "))

myls=[mykor, myeng, mymath, mysce, myhis]
mysum=sum(myls)
myavg=int(mysum/5)

alls = [85, 82, 89, 75, 94] 
subls = ["국어", "영어", "수학", "과학", "역사"]
allsum = sum(alls)
allavg = int(allsum/5)

print("--------------------------------------------------------")
print(f"총점: {mysum}({mysum-allsum}), 평균: {myavg}({myavg-allavg})")
for i in range(4):
    print(f"{subls[i]}점수: {myls[i]}({myls[i]-alls[i]})",end=", ")
print(f"{subls[4]}점수: {myls[4]}({myls[4]-alls[4]})")
print("--------------------------------------------------------")
subls+=["총점","평균"]
myls+=[mysum, myavg]
alls+=[allsum, allavg]
for i in range(7):
    if myls[i]-alls[i] >= 0:
        print(f"{subls[i]} 편차: ","+"*(myls[i]-alls[i]),f"({myls[i]-alls[i]})")
    else :
        print(f"{subls[i]} 편차: ","-"*(alls[i]-myls[i]),f"(-{alls[i]-myls[i]})")
print("--------------------------------------------------------")
```

```python
import random

comNumber = random.randint(1,1000)

userNumber = int(input('1에서 1000까지의 정수 입력: '))
cnt=1 #시도 횟수

while userNumber != comNumber:
    if userNumber < comNumber :
        print("난수가 크다!") 
    else : 
        print("난수가 작다!")
    cnt += 1
    userNumber = int(input('1에서 1000까지의 정수 입력: '))
print("빙고!")
print("난수: {}, 시도 횟수: {}".format(comNumber, cnt))
```

```python
num = int(input("정수 입력: "))
numsum, oddsum, evensum, numpect = 0, 0, 0, 1

for a in range(1,num+1):
    numsum +=a
print("합 결과: {}".format(numsum))

for b in range (1, num+1, 2):
    oddsum +=b
print("홀수 합 결과: {}".format(oddsum))

for c in range (0, num+1, 2):
    evensum +=c
print("짝수 합 결과: {}".format(evensum))

for d in range (1, num+1):
    numpect *=d
print("팩토리얼 결과: {}".format(format(numpect, ",")))
```

```python
n1 = int(input("GearA 톱니수 입력: "))
n2 = int(input("GearB 톱니수 입력: "))
GearA, GearB, cntA, cntB = 0, 0, 0, 0
while n2 % n1 != 0:
    GearB += n2
    cntB += 1
    GearA += n1
    cntA += 1
    print("gearA: {},gearB: {}".format(GearA, GearB))

print("최초 만나는 톱니수(최소공배수): {}톱니".format(GearB))
print("gearA 회전수: {}회전".format(cntA))
print("gearB 회전수: {}회전".format(cntB))
```
